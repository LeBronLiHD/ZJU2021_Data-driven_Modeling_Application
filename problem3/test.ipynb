```python
# 加载数据集
import numpy as np
import preprocess
from sklearn import linear_model
import pandas as pd 
import numpy as np
a=np.loadtxt('BLK_trainX.csv',dtype=np.float32,delimiter=',')
b=np.loadtxt('BLK_trainY.csv',dtype=np.float32)
c=np.loadtxt('BLK_testX.csv',dtype=np.float32,delimiter=',')
a=preprocess.data_cleaning(a)
c=preprocess.data_cleaning(c)
a=preprocess.smaller_than_zero(a)
c=preprocess.smaller_than_zero(c)
np.savetxt('train_input.csv',a)
np.savetxt('test_input.csv',c)
```

```python
#定义线性回归模型
def test_LinearRegrssion():
    import numpy as np
    from sklearn import linear_model
    import pandas as pd 
    import numpy as np
    x_train=np.loadtxt('train_input.csv',dtype=np.float32)
    y_train=np.loadtxt('BLK_trainY.csv',dtype=np.float32)
    x_test=np.loadtxt('test_input.csv',dtype=np.float32)
    x_test=np.loadtxt('test_input.csv',dtype=np.float32)
    print(x_train.shape)
    regr = linear_model.LinearRegression()
    regr.fit(x_train,y_train)#训练数据
    print('Coefficients:%s,intercept %.2f' % (regr.coef_, regr.intercept_))#权重向量即为每个特征的相关系数
    y_test=regr.predict(x_test)
    data = pd.DataFrame(y_test)
    writer = pd.ExcelWriter('test.xlsx')
    data.to_excel(writer, 'page_1', float_format='%.5f')		# ‘page_1’是写入excel的sheet名
    writer.save()
    writer.close()
test_LinearRegrssion()
```
