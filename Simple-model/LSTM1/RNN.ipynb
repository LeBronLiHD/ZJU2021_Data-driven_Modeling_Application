```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

train_input=np.loadtxt('train_modify.txt',dtype=np.float32)
output=np.loadtxt('output_modify.txt',dtype=np.float32)

train_input = pd.DataFrame(train_input)
data_train = scaler.fit_transform(train_input)
data_train=pd.DataFrame(data_train)

output = pd.DataFrame(output)
data_output = scaler.fit_transform(output)
data_output=pd.DataFrame(data_output)
```

```python
#训练集
data_train.describe()
```

```python
#训练集输出
data_output.describe()
```

```python
#测试集
test_input=np.loadtxt('test_modify.txt',dtype=np.float32)
data_test=pd.DataFrame(test_input)
data_test.describe()

# scaler = MinMaxScaler(feature_range=(0, 1))
# result = scaler.fit_transform(data)
```

```python
from sklearn.model_selection import train_test_split
#data_train(全部数据) output(全部目标值)
#X_train训练集(全部特征) Y_train训练集的目标值
#test同理。
#这里没有设置random
X_train, X_test, Y_train, Y_test = \
    train_test_split(data_train,
                     data_output,
                     test_size=0.2,
                     shuffle=False,
                     random_state=1) #这里训练集75%:测试集25%
X_train.shape
X_train_copy = X_train
```

```python
# 简单的多层RNN
import matplotlib.pyplot as plt
from tensorflow import keras
model = keras.models.Sequential([
    keras.layers.SimpleRNN(32,
                           return_sequences=True,
                           input_shape=X_train.shape),
    keras.layers.SimpleRNN(32,return_sequences=True),
    keras.layers.Dense(1)
])
model.compile(loss=keras.losses.mean_squared_error,
              optimizer=keras.optimizers.Adam(0.001),
              metrics=['acc'])
```

```python
#训练模型
X_train = np.expand_dims(X_train,axis=1)

history = model.fit(X_train,Y_train,
                    epochs=50,
                    batch_size=64)

X_test = np.expand_dims(X_test,axis=1)

loss,accuracy = model.evaluate(X_test,Y_test)
print('\ntest loss',loss)
print('accuracy',accuracy)
```

```python
y_pred = model.predict(X_test)
y_pred = np.reshape(y_pred, (y_pred.size,1), order='F')
y_pred = pd.DataFrame(y_pred)
y_pred.index = Y_test.index
fig1, ax1 = plt.subplots(figsize=(8,6))
ax1.plot(y_pred, '--', label='RNN1')
ax1.plot(Y_test, '-',label='Y_test')
ax1.legend(loc='best')
```
